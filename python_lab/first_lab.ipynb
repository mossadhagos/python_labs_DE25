```python
import numpy as np 
import matplotlib.pyplot as plt
from collections import Counter
```

```python
current_key = None
seq_dict = {}

with open("dna_raw.txt", "r") as file:
    for line in file:
        line = line.strip()
        if line.startswith(">"):
            current_key = line[1:]
            seq_dict[current_key] = ""
        else:
            seq_dict[current_key] += line

def count_bases(seq_dict): # defnerat en funktion
    counts_per_key = {}  # anropat en tom dictionery 
    # en for loop 
    for key, seq in seq_dict.items(): # vi har en nyckel med vrde seq 
        counts = Counter(seq) # ny vareable som anvnder countern fr att senare rkna ut 
        counts_per_key[key] = dict(counts) # en varibale som gr igennom en vareje nyckel
    return counts_per_key # reuntern gr att fucnktionen avslutar efetr den har rknat varje vrde
result = count_bases(seq_dict) # detta ger oss resultatet.
print(result)


for key, base_counts in result.items():
    plt.bar(base_counts.keys(), base_counts.values())
    plt.title(f"DNA-frekvens fr {key}")
    plt.xlabel("DNA-bokstav")
    plt.ylabel("Antal")
    plt.show()
```
