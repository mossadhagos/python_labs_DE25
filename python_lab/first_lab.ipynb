#### Första delen av labbet

```python
import matplotlib.pyplot as plt
from collections import Counter
```

- Jag började med att öppna filen och sedan filhantera den. Men jag hade svårt med att omvandla seqvenserna till en lista.
- Jag snackade med Dennis i min klass och han tipsade om att jag borde göra en `Dictionary` och jag började expromentera med det. 
- Jag skappade en tom dictionary och en varibael som heter (current_key).
- Jag frågade även ai om att förklara hur en saker fungerade för att få en bättre förstålese.
- Jag tog hjälp från min vän som har en data sicentes ubliding, som hjälpte med vad som kunde förbätras min kode med genom att ta bort onödiga saker.

```python
# skapar en variabel current_key och sätter den till None (inget värde)
current_key = None

# Jag skappar en dictionary {} kallad för seq_dict.

seq_dict = {}

# open öppnar filen, den öppnar som read mode och som file 
# with hjälper mig att stänga filen när den tar slut så att for loopen inte fortsäter

with open("dna_raw.txt", "r") as file:
    
    # här har vi en for loop som läser filen 
    # Line in file gör att den läser en line åtgången
    
    for line in file:
        
        # Använder line.strip() för att ta bort 
        
        line = line.strip()
       
        # line.starswith(">") kontrollerar om raden börjar med > då har vi en ny seqvens att räkna.
        
        if line.startswith(">"):
           
            # detta tar bort > med slice line[1:] och sparar resten i current_key 
            
            current_key = line[1:]
           
            # Skappar i seq_dict en post med nyckel current_key med en tom string ""
            # Vi förbreder en plats att append:a (lägga till) seqvens rader som kommer efter > 
            
            seq_dict[current_key] = ""
           
            # Om inte nästa rad inte börjar med en > försätta på samma sekvens
        
        else:
            seq_dict[current_key] += line

# Defnerat en funktion
# Denna funktion som kommer räkna antal boktäver i sekvenseran 

def count_bases(seq_dict): 
    # anropat en tom dictionery 
    
    counts_per_key = {}  
    # en for loop 
    
    for key, seq in seq_dict.items(): # vi har en nyckel med värde seq 
        
        # ny vareable som använder countern för att senare räkna ut 
        counts = Counter(seq) 
        
        # en varibale som går igennom en vareje nyckel
       
        counts_per_key[key] = dict(counts) 
        # return gör att fucnktionen avslutar efter den har räknat varje värde
    
    return counts_per_key 
 # detta ger oss resultatet, 

result = count_bases(seq_dict)
print(result)

# Här använder jag mig ut av matplotlib
# Detta kommer göra att alla sekvenser vissas i en fin graf
for key, base_counts in result.items():
    plt.bar(base_counts.keys(), base_counts.values())
    plt.title(f"DNA-frekvens för {key}")
    plt.xlabel("DNA-bokstav")
    plt.ylabel("Antal")
    plt.show()
```
